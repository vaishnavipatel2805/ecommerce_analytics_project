## Dataset Overview

The dataset consists of **9 relational tables** covering the complete order lifecycle:

| Table | Rows | Key |
|---|---|---|
| `orders` | 99,441 | order_id |
| `order_items` | 112,650 | order_id, product_id |
| `customers` | 99,441 | customer_id |
| `products` | 32,951 | product_id |
| `sellers` | 3,095 | seller_id |
| `reviews` | 99,224 | review_id |
| `payments` | 103,886 | order_id |
| `geolocation` | 1,000,163 | zip_code_prefix |
| `category_translation` | 71 | product_category_name |

As of now we will use only **5 relational tables** :

1. orders
2. customers
3. items
4. payments
5. reviews

## Dataset Structure & Relationships

Before any analysis, we need to understand how the tables relate to each other.
This is a **star-schema style** relational dataset — `orders` sits at the center,
and all other tables join to it through a primary key.

```
orders ──── customers        (customer_id)
orders ──── order_items      (order_id)
orders ──── reviews          (order_id)
orders ──── payments         (order_id)
order_items ── products      (product_id)
order_items ── sellers       (seller_id)
products ─── category_tr    (product_category_name)
```


# 3. Data Cleaning & Merging

### converting datetime cols

````python
date_cols = [
    "order_purchase_timestamp", "order_approved_at",
    "order_delivered_carrier_date", "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in date_cols:
    orders[col] = pd.to_datetime(orders[col], errors="coerce")

print("✅ Date columns converted")
orders[date_cols[:5]].dtypes
````

````python
orders.info()
````

````python
reviews["review_creation_date"] = pd.to_datetime(reviews["review_creation_date"], errors="coerce")
reviews["review_answer_timestamp"] = pd.to_datetime(reviews["review_answer_timestamp"], errors="coerce")
reviews.info()
````

````python
items["shipping_limit_date"] = pd.to_datetime(items["shipping_limit_date"], errors="coerce")
````

````python
items.info()
````

### creating master dataset

````python
df = orders.merge(customers, on='customer_id', how='left')
df = df.merge(items, on='order_id', how='left')
df = df.merge(payments, on='order_id', how='left')
df = df.merge(reviews, on='order_id', how='left')
````

````python
df.head()
````

````python
df.info()
````

````python
missing_report = pd.DataFrame({
    "missing_count": df.isnull().sum(),
    "missing_pct": (df.isnull().mean()*100).round(2)                 
})
missing_report = missing_report[missing_report["missing_count"]>0].sort_values(by = "missing_count",ascending = False)
missing_report
````

# 4. Feature Engineering

````python
df["delivery_days"] = (df["order_delivered_customer_date"] - df["order_purchase_timestamp"]).dt.days
````

````python
df["delivery_delay"] = ( df["order_delivered_customer_date"] - df["order_estimated_delivery_date"] ).dt.days
````

````python
df['total_price'] = df['price'] + df['freight_value']
````

````python
df["order_month"] = df["order_purchase_timestamp"].dt.month_name()
````
